```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
from matplotlib.dates import AutoDateLocator, DateFormatter  

plt_df = pd.read_csv('Result/cws_process.csv')

plt.title('CRF Train F1 scroe')
plt.xlabel('Train Epoch')
plt.ylabel('F1 score(%)')
params = {
            'axes.labelsize': '16',
            'xtick.labelsize': '16',
            'ytick.labelsize': '13',
            'lines.linewidth': '2',
            'legend.fontsize': '20',
            'figure.figsize': '20, 14'  # set figure size
}
pylab.rcParams.update(params)

print(len(plt_df['id']))
plt.plot(list(range(len(plt_df['id']))), plt_df['train'],'r', label='Train')
plt.plot(list(range(len(plt_df['id']))), plt_df['dev'],'b',label='Dev')
plt.xticks(list(range(len(plt_df['id']))), plt_df['id'], rotation=0)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab

plt_df = pd.read_csv('Result/cws_bilstmcrf.csv')

plt.title('BiLSTM CRF Train F1 scroe')
plt.xlabel('Train Epoch')
plt.ylabel('F1 score(%)')
params = {
#             'axes.labelsize': '16',
#             'xtick.labelsize': '20',
#             'ytick.labelsize': '13',
#             'lines.linewidth': '2',
            'legend.fontsize': '5',
            'figure.figsize': '25, 17.5'  # set figure size
}
pylab.rcParams.update(params)

sample = [ii for ii in range(len(plt_df['id'])) if not ii % 2]
train = [plt_df['train'][ii] for ii in sample ]
dev = [plt_df['dev'][ii] for ii in sample ]
idx = [plt_df['id'][ii] for ii in sample ]
plt.plot(sample, train,'r', label='Train')
plt.plot(sample, dev,'b',label='Dev')
plt.xticks(sample, idx)
plt.show()
```
